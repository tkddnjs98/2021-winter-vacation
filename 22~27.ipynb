```python
scores=[10, 20, 30, 40]
n_scores=len(scores)
print(n_scores)
```

```python
#리스트의 합을 구하기
```

```python
# for 문 사용
sum1=0
for i in scores: #i가 scores에서 인자를 받아 하나씩 더하기
    sum1+=i
n_student=len(scores)
mean=sum1/n_student
print(mean)

sum2=0
for i in range(4):
    sum2+=scores[i]
mean2=sum2/n_student
print(mean2)
```

1-25 list의 variance 구하기

```python
scores=[10,20,30,40]
n_student=len(scores)
score_mean=(scores[0]+scores[1]+scores[2]+scores[3])/n_student
score_mean
```

```python
square_of_mean=score_mean**2
mean_of_square=(scores[0]**2+scores[1]**2+scores[2]**2+scores[3]**2)/n_student
variance=mean_of_square-square_of_mean
std=variance**0.5
print("표준편차:",std)
```

```python
scores=[10,20,30]
n_student=len(scores)
mean=(scores[0]+scores[1]+scores[2])/n_student
square_of_mean=mean**2
print(square_of_mean)
mean_of_square=(scores[0]**2+scores[1]**2+scores[2]**2)/n_student
variance=mean_of_square-square_of_mean
print(variance)
std=variance**0.5
print("표준편차:",std)
```

1-27 리스트의 표준화

```python
scores=[10, 20, 30]
#before_standardization
n_student=len(scores)
mean=(scores[0]+scores[1]+scores[2])/n_student
s_mean=mean**2
mean_s=(scores[0]**2+scores[1]**2+scores[2]**2)/n_student
variance=mean_s-s_mean
print(variance)
std=variance**0.5
print("표준편차:",std)
#after std
scores[0]-=mean
scores[1]-=mean
scores[2]-=mean
scores[0]/=std
scores[1]/=std
scores[2]/=std
mean=(scores[0]+scores[1]+scores[2])/n_student
print(mean)
s_mean=mean**2
mean_s=(scores[0]**2+scores[1]**2+scores[2]**2)/n_student
variance=mean_s-s_mean
print("std:",variance**0.5)
```
