분산,표준편차 구하기

제곱의 평균-평균의 제곱=분산 //
표준 편차는 분산에 power을 0.5곱해서 구현

```python
score1,score2,score3= 10,20,30
n_student=3
#평균 구하기
mean=(score1+score2+score3)/n_student
print("평균:",mean)
#평균의 제곱
square_of_mean=mean**2
#제곱의 평균
mean_of_square=(score1**2+score2**2+score3**2)/n_student
print(mean_of_square)
variance=mean_of_square-square_of_mean
print("분산:{}".format(variance))
std=variance**0.5
print("표준편차:{}".format(std))
```

표준화: 평균을 뺴고 표준 편차를 나누기
평균은 0, 표준 편차는 1로 만들기

```python
score1,score2,score3=10,20,30
print("학생당 점수:",score1,score2,score3)
```

```python
mean=(score1+score2+score3)/n_student
print("평균:",mean)
square_of_mean= mean**2
mean_of_square=(score1**2+score2**2+score3**2)/n_student
print("제곱의 평균:",mean_of_square)
variance=mean_of_square-square_of_mean
print("분산",variance)
std=variance**0.5
print("표준편차:",std)
#표준화
score1_n=(score1-mean)/std
score2_n=(score2-mean)/std
score3_n=(score3-mean)/std
mean_n=(score1_n+score2_n+score3_n)/n_student
print("표준화한 평균:",mean_n)
variance_n=(score1_n**2+score2_n**2+score3_n**2)/n_student-mean_n**2
print("표준화한 표준편차:",variance_n**0.5)
```
