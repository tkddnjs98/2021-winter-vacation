89 리스트를 예쁘게

```python
vectors=[[1, 11, 21],
         [2, 12, 22],
         [3, 13, 23],
         [4, 14, 24]]
#dim_product는 곱이라서 1로 설정
'''
v1 vector:[1, 2, 3, 4]
v2 vector:[11, 12, 13, 14]
v3 vector:[21, 22, 23, 24]
'''
```

하다마드 구현 줄에 하나씩 접근, 차원 내의 원소끼리 곱하기 맨위는 1차원 아래는 2차원 그다음은 3차원 마지막은 4차원

```python
h_prod=list()
for vector_dim in vectors:
    dim_prod=1 #곱을 1로 초기화
    for vector in vector_dim:
        dim_prod *=vector
    h_prod.append(dim_prod)
print(h_prod)
```

Dot Product

```python
vectors=[[1, 11],
         [2, 12],
         [3, 13],
         [4, 14]]
dot_prod=0
for n_dims in range(len(vectors)):
    dot_prod+=vectors[n_dims][0]*vectors[n_dims][1]
print(dot_prod)
```
