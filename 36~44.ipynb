1-42 for loop로 list의 원소 접근하기

```python
#for loop와 index로 접속하기
scores = [10, 20, 30, 40, 50]
print("===원래 방식===")
print(scores[0])
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])
print("===새로운 방식===")
#배우는 방식
scores= [10, 20, 30, 40, 50]
for idx in range(len(scores)): #for idx in range(5)와 같다.
    print(idx, scores[idx])
```

```python
scores1=[10, 20, 30, 40, 50, 60]
print("===직접접근===")
for score in scores1:
    print(score)
print("===인덱스로 접근===")
for idx in range(len(scores1)):
    print(scores1[idx])
```

1-44 리스트3개의 원소끼리 동시에 더하기 얘도 인덱스 사용

```python
list1=[1, 2, 3, 4]
list2=[10, 20, 30, 40]
list3=[]
for idx in range(len(list1)):
    print(list1[idx], list2[idx])
    print(list1[idx]+list2[idx])
    list3.append(list1[idx]+list2[idx])
print(list3)
```
