1-46 mean_subtractions #평균 0으로 만들기

```python
scores= [10, 20, 30, 40, 50, 60]
print(scores)
score_sum=0
for i in range(len(scores)):
    score_sum +=scores[i]
mean= score_sum/len(scores)
print(mean)
score_sum1=0
for i in range(len(scores)):
    scores[i] -=mean
print(scores)
for j in range(len(scores)):
    score_sum1+= scores[j]
print(score_sum1/len(scores))
#원본 값이 사라진다.
```

```python
scores= [10, 20, 30, 40, 50, 60]
sum_score=0
for score in scores:
    sum_score +=score
mean=sum_score/len(scores)
list3=[]
for score in scores:
    list3.append(score-mean)
print(list3)
sum_list3=0
for score in list3:
    sum_list3 +=score
mean2=sum_list3/len(scores)
print(mean2)
#다른 변수로 이동하기 때문에 원래 리스트 재사용 가능
```

```python
scores=[10,20,30]
score_sum=0
for score in scores:
    score_sum +=score
mean= score_sum/len(scores)
print(mean)
list3=[]
for score in scores:
    list3.append(score-mean)
print(list3)
sum_scores2=0
for score2 in list3:
    sum_scores2 +=score2
print(sum_scores2/len(list3))
```

```python
scores=[10,20,30]
list3=[]
score_sum=0
for i in range(len(scores)):
    score_sum +=scores[i]
mean1=score_sum/len(scores)
for i in range(len(scores)):
    list3.append(scores[i]-mean1)
print(list3)
score_sum2=0
for score in list3:
    score_sum2 +=score
print(score_sum2/len(list3))
```

1-47 2개 리스트의 평균을 들고 다니기
[0,1] 0부분에는 수학의 합 1부분에는 영어의 합이 들어간다.

```python
math_scores= [40, 60, 80]
english_scores= [30, 40, 50]
n_classes=2
scores_sum=[0 for i in range(n_classes)]
for idx in range(len(math_scores)):
    scores_sum[0] += math_scores[idx]
    scores_sum[1] += english_scores[idx]
score_mean=[]
print(scores_sum)
for idx in range(len(scores_sum)):
    score_mean1=scores_sum[idx]/len(math_scores)
    score_mean.append(score_mean1)
print(score_mean)
```

1-48 2개의 리스트의 mean subtraction

```python
math_scores=[40, 60, 80]
english_scores=[30, 40, 50]
n_classes
sum_scores=[0]*(n_classes)
for i in range(len(english_scores)):
    sum_scores[0] +=math_scores[i]
    sum_scores[1] +=english_scores[i]
print(sum_scores)
#평균 구하기
mean_scores=[]
mean_scores.append(sum_scores[0]/len(math_scores))
mean_scores.append(sum_scores[1]/len(english_scores))
print(mean_scores)
mean2=[0]*2
minus_math=[]
minus_english=[]
for i in range(len(math_scores)):
    minus_math.append(math_scores[i]-mean_scores[0])
    minus_english.append(english_scores[i]-mean_scores[1])
print(minus_math)
print(minus_english)
for i in range(len(math_scores)):
    mean2[0]+=(math_scores[i]-mean_scores[0])
    mean2[1]+=(english_scores[i]-mean_scores[1])
print(mean2[0]/len(math_scores))
print(mean2[1]/len(math_scores))
```
