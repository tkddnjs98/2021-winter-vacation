1-63 분, 초를 만들기

```python
seconds =200
if seconds >60:
    minutes = seconds //60
    seconds -=minutes*60
else:  #60초보다 같거나 작은 경우
    minutes = 0
print(minutes, 'min',seconds,'sec')
```

```python
#재귀 알고리즘을 이용
m=0
def clock(seconds):
    global m
    if seconds >= 60:
        m +=1
        return clock(seconds-60)
    else:
        return m,seconds
    print(m)
```

```python
clock(10000)
```

1-64 초를 시 분 초로 변환

```python
seconds =5000
hour =0
minutes =0
if seconds >= 3600:
    hour= seconds//3600
    seconds -=hour*3600
else: hour = 0


if seconds > 60:
    minutes= seconds //60
    seconds -= minutes*60
else: minutes =0

print(hour, minutes, seconds)
```

1-67 절대 평가를 해보기

```python
scores = [10, 20, 30, 40, 50, 60, 70]
for idx in range(len(scores)):
    if scores[idx] >= 60:
        print("A")
    elif scores[idx] >=55:
        print("B")
    elif scores[idx] >= 50:
        print("C")
    else:
        print("F")
```
