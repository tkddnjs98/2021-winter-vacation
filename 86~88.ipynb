```python
#점수 list에 접근하기 위해 학생 인덱스 필요
#과목에 접근 하기위한 과목 인덱스 이용
```

```python
#과목 평균 구하기
#평균을 구하기
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
mean_list=[]
sum_list=[]
for _ in range(len(scores[0])):
    sum_list.append(0)
for student_score in scores:
    for idx in range(len(student_score)):
        sum_list[idx]+=student_score[idx]
print(sum_list)
for idx in range(len(sum_list)):
    mean_list.append(sum_list[idx]/len(scores))
print(mean_list)
#평균을 빼보기

for student_idx in range (len(scores)):
    for class_idx in range (len(scores[0])):
        scores[student_idx][class_idx] -=mean_list[class_idx]
print(scores)
#다시 평균 구해보기
sub_sum=[]
for _ in range(len(scores[0])):
    sub_sum.append(0)
for score in scores:
    for idx in range(len(scores[0])):
        sub_sum[idx] +=(score[idx])
print(sub_sum)
sub_mean=[]
for idx in range(len(scores[0])):
    sub_mean.append(sub_sum[idx]/len(score))
print(sub_mean)
```

분산과 표준 편차를 구해보기 분산: 제곱끼리의 평균-평균의 제곱

```python
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
class_mean=[]
n_class=len(scores[0])
for _ in range(n_class):
    class_mean.append(0)
for score in scores:
    for idx in range(len(score)):
        class_mean[idx] +=score[idx]
for idx in range(n_class):
    class_mean[idx] /=len(scores)
print(class_mean)

variation=[]
power_sum=[]
for idx in range(len(class_mean)):
    class_mean[idx]**=2
for _ in range(n_class):
    power_sum.append(0)
    variation.append(0)
for score in scores:
    for idx in range(n_class):
        power_sum[idx] +=score[idx]**2
for idx in range(n_class):
    power_sum[idx]/= len(scores) #분산에서는 제곱의 합은 학생의 수만큼 나누어야 한다. 
for idx in range(n_class):
    variation[idx]+=power_sum[idx]-class_mean[idx]
print(variation)
std_list=[]
for idx in range(n_class):
    std_list.append(variation[idx]**0.5)
print(std_list)
```

```python
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
n_classes=len(scores[0])
n_student=len(scores)
power_mean=[]
square_sum=[]
mean_list=[]
sum_list=[]
variance_list=[]
for _ in range(n_classes):
    sum_list.append(0)
    square_sum.append(0)
for score in scores:
    for idx in range(n_classes):
        sum_list[idx] +=score[idx]
        square_sum[idx] +=score[idx]**2
for idx in range(n_classes):
    mean_list.append(sum_list[idx]/n_student)
    square_sum[idx]/=n_student
for idx in range(n_classes):
    power_mean.append(mean_list[idx]**2)
for idx in range(n_classes):
    variance_list.append(square_sum[idx]-power_mean[idx])
for idx in range(n_classes):
    variance_list[idx] **=0.5
#표준화
for student_idx in range(n_student):
    for class_idx in range(n_classes):
        score=scores[student_idx][class_idx]
        mean=mean_list[class_idx]
        std=variance_list[class_idx]
        scores[student_idx][class_idx]=(score-mean)/std        
print(scores)

#표준화 이후 평균과 분산 구하기
n_classes=len(scores[0])
n_student=len(scores)
power_mean=[]
square_sum=[]
mean_list=[]
sum_list=[]
variance_list=[]
for _ in range(n_classes):
    sum_list.append(0)
    square_sum.append(0)
for score in scores:
    for idx in range(n_classes):
        sum_list[idx] +=score[idx]
        square_sum[idx] +=score[idx]**2
for idx in range(n_classes):
    mean_list.append(sum_list[idx]/n_student)
    square_sum[idx]/=n_student
for idx in range(n_classes):
    power_mean.append(mean_list[idx]**2)
for idx in range(n_classes):
    variance_list.append(square_sum[idx]-power_mean[idx])
for idx in range(n_classes):
    variance_list[idx] **=0.5
print(variance_list)
print(mean_list)
```
