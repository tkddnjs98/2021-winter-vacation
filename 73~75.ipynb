1-73 최대와 최소값을 가정하지 않고 최대와 최소값을 구할 수 있을까?


```python
#방법 1: 0번 인덱스의 값을 최대, 1번 인덱스의 값을 최소라 가정하고 해보기!
scores=[120, 130, 150, 180]
M,m=scores[0],scores[0]
for score in scores:
    if score>M:
        M=score
    if score<m:
        m=score
print("Max value:",M)
print("Min value",m)
```

```python
#방법 2: None으로 초기화 하기
scores=[-20, 60, 40, 70, 120]
M,n=None, None
for score in scores:
    if M==None or M<score: #or은 둘 중 하나만 성립하면 되고 왼쪽에 있는 조건문이 먼저 실행 되기 때문에 None여부를 왼쪽에 위치해야 한다.
        M=score
    if n==None or n>score:
        n=score
print("Max value:",M)
print("Min value",n)
```

1-74 Normalization을 해보기
Max-Min normalization
(x-최소)/(최대-최소)로 standadization을 수행
100~200사이 데이터 가정: 0에서 100사이 값으로 옮길 수 있으며 여기서 100을 나누면? 0에서 1사이 값이 나온다!


```python
scores=[-20, 60, 40, 70, 120]
#최대랑 최소구하기
M,m=None,None
for score in scores:
    if M==None or M<score:
        M=score
    if m==None or m>score:
        m=score
#노멀라이즈 하기
for idx in range(len(scores)):
    scores[idx]=(scores[idx]-m)/(M-m)
print(scores)
M,m=None,None
for score in scores:
    if M==None or M<score:
        M=score
    if m==None or m>score:
        m=score
print("최대값:",M,"최소값",m)
```

1-75최대값의 인덱스와 최소값의 인덱스를 구하기!
이거 구할 때는 인덱스를 저장할 변수, 인덱스로 리스트를 접근!!

```python
scores=[60, -20, 40, 70, 120]
Max_idx,Min_idx= 0, 0
m,M=None, None
for idx in range(len(scores)):
    if M==None or M<scores[idx]:
        Max_idx=idx
        M=scores[idx]
    if m==None or m>scores[idx]:
        Min_idx=idx
        m=scores[idx]
print("최대 값과 최대의 인덱스:",M,Max_idx,"\n최소 값과 최소의 인덱스",m,Min_idx)
```
