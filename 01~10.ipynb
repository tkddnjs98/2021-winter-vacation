1-9 변수를 각각 증가

```python
score1+=10
score2+=10
score3+=10
```

```python
score_mean=(score1+score2+score3)/n_student
```

```python
print(score_mean) #변수가 모두10씩 증가하자, 평균도 10이 증가한다.
#선형성이 유지
```

```python
score1*=5
score2*=5
score3*=5
```

```python
score_mean=(score1+score2+score3)/3
```

```python
print(score_mean) #모든 변수에 5씩 곱하자, 평균이 5배가 됨
```
