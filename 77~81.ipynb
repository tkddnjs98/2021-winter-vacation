1-77 Accuracy구하기
몇개가 맞았는지 측정

```python
import numpy as np
predictions=[0, 1, 0, 2, 1, 2, 0]
labels=[1, 1, 0, 0, 1, 2, 1]
n_data=len(predictions)
n_correct=0
n_count=0
for idx in range(len(predictions)):
    if predictions[idx]==labels[idx]:
        n_correct+=1 #같으면 맞은개수 1개 올라감 (이거 파이토치에 활용)
        n_count +=1
    else: 
        n_count +=1
accuracy=n_correct/n_count
print(np.round(accuracy,2),"%")
```

1-78 Confusion Matrix 구현 #다시 해보기

vector 별로 정확도를 구해보기
0,1,2 3가지중 어느 항목이 제일 정확도가 높은지 생각하기
각각 class별 정확도 구하기
1. 제일 큰 class구하기. 최대값이 maximum class이다.
2. class의 개수를 따지는 것이기 때문에 제일 큰 값에 1을 꼭 더한다. 보통 개수 3개라 하면 0포함 0,1,2 가 되니까 3번 하려고
3. maxim_class의 횟수를 반복하여. class count, correct count, confusion_vec의 리스트를 생성.
4. 하나씩 값에 접근을 한다. 
5. class_count: class의 개수 correct_count 맞은 개수
6. 인덱스에 해당하는 class_count/correct_count를 하여 정확도 구하기.

```python
predictions=[0, 1, 0, 2, 1, 2, 0]
labels=[1, 1, 0, 0, 1, 2, 1]
max_index,M=0,predictions[0]
for idx in range(len(predictions)):
    if M<predictions[idx]:
        M=predictions[idx]
print(M)
```

 class의 개수: 2+1=3 (0, 1, 2)라서

```python
num_classes=M+1
correct_classes, num_of_classes, accuracy_vector=[],[],[]
for i in range (num_classes):
    correct_classes.append(0)
    num_of_classes.append(0)
    accuracy_vector.append(None)
print(correct_classes, num_of_classes,accuracy_vector)
for idx in range(len(predictions)):
    if predictions[idx]==labels[idx]:
        num_of_classes[labels[idx]]+=1
        correct_classes[labels[idx]]+=1
    else:
        num_of_classes[labels[idx]]+=1
print(correct_classes)
print(num_of_classes)
```

```python
for idx in range(num_classes):
    accuracy_vector[idx]=correct_classes[idx]/num_of_classes[idx]
```

```python
scores=[50, 20, 30, 40, 10, 50, 70, 80, 90, 20, 30]
cutoffs=[0, 20, 40, 60, 80]
histogram=[0]*5
for score in scores:
    if score>cutoffs[4]:
        histogram[4] +=1
    elif score>cutoffs[3]:
        histogram[3] +=1
    elif score>cutoffs[2]:
        histogram[2] +=1
    elif score>cutoffs[1]:
        histogram[1] +=1
    elif score>cutoffs[0]:
        histogram[0] +=1
    else:
        pass #연산 안하고 넘기기
print(histogram)
    
    
```

절대값 구하기

```python
numbers=[-2, 2, -1, 3, -4, 9]
abs_numbers=list()
for number in numbers:
    if number <0:
        abs_numbers.append(-number)
    else:
        abs_numbers.append(number)
print(abs_numbers)
```

L1 norm 이용, 맨하튼 distance 이용 1차원의 차이의 절대값이용

```python
v1=[1, 3, 5, 2, 1, 5, 2]
v2=[2, 3, 1, 5, 2, 1, 3]
m_distance=0
for idx in range (len(v1)):
    sub=v1[idx]-v2[idx]
    if sub <0:
        m_distance +=-sub
    if sub>0:
        m_distance +=sub
print(m_distance)
```
