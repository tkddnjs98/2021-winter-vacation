분산과 표준편차, standardization

1-49 복습: 분산은 제곱의 평균에서 평균의 제곱을 뺀다.

```python
scores=[10, 20, 30]
score_sum=0
for i in range(len(scores)):
    score_sum +=scores[i]
mean= score_sum/len(scores)
print("평균:", mean)
square_of_mean= mean**2
mean_of_square=0
for i in range(len(scores)):
    mean_of_square +=scores[i]**2
mean_of_square /= len(scores)

variance=(mean_of_square-square_of_mean)
print("분산:",variance)
std= variance **0.5
print("표준편차:",std)
```

1-50 표준화

```python
scores= [10, 20, 30]
sum_of_scores=0
mean_of_square=0
variance=0
std=0
print("===before standardization===")
for i in range(len(scores)):
    sum_of_scores +=scores[i]
    mean_of_square +=scores[i]**2
mean= sum_of_scores/len(scores)
mean_of_square /=len(scores)
square_of_mean= mean**2
variance=mean_of_square-square_of_mean
print("분산:", variance)
std=variance **0.5
print("표준편차:", std)
#standardization
list3=[]
for i in range(len(scores)):
    list3.append((scores[i]-mean)/std)


print("===after standardization===")
print(list3)
sum_of_standard=0
sum_of_square=0
for i in range(len(list3)):
    sum_of_standard +=list3[i]
    sum_of_square +=list3[i]**2
mean2=sum_of_standard/len(list3)
square_of_mean2= mean2**2
variance2= sum_of_square/len(list3)-square_of_mean2
std2=variance2**0.5
print("표준화한 평균:", mean2,"표준화한 표준편차:" ,std2)
```

수학과 영어 2개의 리스트로 분산 구하기

```python
math_scores, english_scores = [50, 60, 70], [30, 40, 50]
n_student=len(math_scores)
sum_of_math, sum_of_english= 0,0
sum_of_math_square, sum_of_english_square = 0,0
for i in range(n_student):
    sum_of_math+=math_scores[i]
    sum_of_math_square+=math_scores[i]**2
    sum_of_english+=english_scores[i]
    sum_of_english_square+=english_scores[i]**2
mean_of_math=sum_of_math/n_student
mean_of_english=sum_of_english/n_student
mean_of_math **=2
mean_of_english **=2
print(mean_of_english)
variance_of_math=sum_of_math_square/n_student-mean_of_math
variance_of_english=sum_of_english_square/n_student-mean_of_english
print("수학 분산:",variance_of_math,"영어 분산:",variance_of_english)
print("수학 표준 편차:",variance_of_math**0.5, 
      "영어 표준 편차:",variance_of_english**0.5)
```
