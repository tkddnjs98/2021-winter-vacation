1-19 유클리드 거리(L2)
n차원 공간산에서 v와 u가 있다고 가정
거리를 표현 (v1,u1) (v2,u2)        
루트((v1-v2)^2+(u1-u2)^2) 빼서 제곱하고 합친다음 루트

```python
x1,y1,z1=1,2,3
x2,y2,z2=3,4,5
u_distance=((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)**0.5
print(u_distance)

u_distance=((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)
u_distance**=0.5
print(u_distance)
```

1-21 mean squared error
제곱합의 평균

```python
y1,y2,y3=10,25,40
pred1,pred2,pred3=10,20,30
n_data=3
```

```python
error1=(y1-pred1)**2
error2=(y2-pred2)**2
error3=(y3-pred3)**2
```

```python
mse=(error1+error2+error3)/n_data
print("mean squared error:",mse)
```
