1-68 리스트에 적용해서 학점 매기기

```python
scores = [10, 20, 30, 40, 50, 60, 70]
for idx in range(len(scores)):
    if scores[idx] >= 60:
        print("A")
    elif scores[idx] >=55:
        print("B")
    elif scores[idx] >= 50:
        print("C")
    else:
        print("F")
```

```python
scores=[10, 20, 30, 40, 50, 60, 70]
cutline=30
for score in scores:
    if score>cutline:
        print("Pass")
    else:
        print("Fail")
```

1-69 리스트에 학점을 append하기

```python
scores=[70, 60, 50, 40, 30, 20, 10, 20, 30, 35, 40 ]
grades=[]
for score in scores:
    if score>60:
        grades.append("A")
    elif score>30:
        grades.append("B")
    elif score>20:
        grades.append("C")
    else:
        grades.append("F")
print(grades)
```

50기준으로 합격, 불합격자 학생들을 나누고 각각 평균 구하기
합격점수 +누적과 합격 학생수 +1하고, 불합격점수+누적하고 불합격 학생 +1을 한다.

```python
scores=[70, 60, 50, 40, 30, 20, 10, 20, 30, 35, 40 ]
p_c,n_pc= 0,0
p_score, np_score=0,0
for score in scores:
    if score>50:
        p_c +=1
        p_score +=score
    else:
        n_pc +=1
        np_score +=score
print("Pass 평균:",p_score/p_c)
print("Non-Pass 평균:",np_score/n_pc)
        
```

```python
scores=[20, 50, 10, 60, 90]
p_c,n_pc= 0,0
p_score, np_score=0,0
for score in scores:
    if score>50:
        p_c +=1
        p_score +=score
    else:
        n_pc +=1
        np_score +=score
print("Pass 평균:",p_score/p_c)
print("Non-Pass 평균:",np_score/n_pc)
```

넘버 리스트를 추가 그리고 소수를 추가
number의 나머지가 0이면 짝수, 홀수 
그럼 나머지는? 정수 아니라고 정함.

```python
import numpy as np
numbers=np.arange(1,21,1) #1부터 20인 수 생성
numbers=list(numbers)
numbers.append(3.14)
print(numbers)
for number in numbers:
    if number%2==1:
        print("odd")
    elif number %2 ==0:
        print("even")
    else:
        print("Not a number")
```

```python
numbers=list()
for i in range(20):
    numbers.append(i)
numbers.append(3.14)
print(numbers)
for numbers in numbers:
    if numbers%2 ==0:
        print("even")
    elif numbers%2 ==1:
        print("odd")
    else:
        print("Not a integer")
```

N의 배수에 대한 합

```python
numbers=np.arange(1,100,1)
print(numbers)
sum_num=0
for idx in range(len(numbers)):
    if numbers[idx]%3==0:
        sum_num+=numbers[idx]
print(sum_num)
```

```python
multiple_of=3
numbers=list()
for num in range(100):
    numbers.append(num)
sum_of_multiple=0
for number in numbers:
    if number%multiple_of==0:
        sum_of_multiple+=number
print("3의 배수의 합:",sum_of_multiple)
```

최대값과 최소값을 구현 
최소 값<큰 값이면 최대 값이 갱신이 된다.

```python
scores=[60, 40, 70, 20, 30]
max_value=0
min_value=100
for score in scores:
    if score>max_value:
        max_value=score
    #다른 if문 작동
    if score<min_value:
        min_value=score
print(max_value, min_value)
```
