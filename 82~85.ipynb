1-84 학생별 평균 구하기
리스트의 개수: 학생의 개수
리스트안의 원소의 개수: 과목의 개수

```python
#내가 해본 풀이. 2차원 리스트의 원소를 직접 접근 해서 2중 루프 이용.
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
n_student=len(scores) #scores의 행이 학생수이다.
mean_students=[]

for i in range(n_student):
    sum_student=0  #매번 초기화
    for j in range(len(scores[0])):
        sum_student+=scores[i][j]
    mean_student=sum_student/len(scores[0]) #더하는 것이 끝나면 평균 구하기.
    mean_students.append(mean_student) #평균 구해서 append 수행.
            
print(mean_students)
        
```

```python
#정답 리스트를 원소에 직접 할당.
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
n_student=len(scores) #4개의 리스트가 있어서 길이가 4이다.
mean_list=[]
for student_scores in scores: #scores에서 점수 리스트를 할당 받는다.
    student_sum=0 #매번 리스트를 받기 때문에 리스트는 매번 0으로 초기화.
    for score in student_scores: #student_scores란  리스트에 저장된 값을 하나씩 score에 할당.
        student_sum+=score 
    mean_list.append(student_sum/len(student_scores)) #반복이 끝나면 평균을 집어 넣기.
print(mean_list)
```

1-85 과목별 평균 구하기
1. 과목 별로 합을 구하기
2. 과목합/학생수로 평균 구하기

```python
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
score_sum_list=[]
score_mean=[]
for j in range(len(scores[0])):
    class_sum=0
    for i in range(len(scores)):
        class_sum+=scores[i][j]
    score_sum_list.append(class_sum) #과목별로 합해서 리스트에 넣기
print(score_sum_list)
for score in score_sum_list:
    score_mean.append(score/4) #과목 합에서 4씩 나눠서 과목별 평균 구하기.
print(score_mean)
```

```python
scores=[[10, 15, 20],[20, 25, 30],[30, 35, 40],[40, 45, 50]]
n_classes=len(scores[0])
n_student=len(scores)
# 먼저 과목 합과 평균 리스트를 생성한다음에 과목의 개수 만큼 0을 넣어서 초기화를 한다.
score_sum_list=[]
score_mean=[]
for _ in range(n_classes):
    score_sum_list.append(0)
print(score_sum_list)
for student_scores in scores:
    for idx in range(n_classes):
        score_sum_list[idx] += student_scores[idx]
#scores에서 학생간 점수 리스트를 받는다.
#0번째 학생의 0,1,2 번째 과목을 누적
#1번째 학생의 0,1,2 번째 과목을 누적
#마지막으로 2번째 학생의 0,1,2번째 과목을 리스트에 각각 누적한다.
#그러면 score_sum_list가 0에서 과목들을 누적한 형태로 바뀐다.
print(score_sum_list)
for idx in range(len(score_sum_list)):
    score_mean.append(score_sum_list[idx]/n_student)
print(score_mean)
```
