1-54 vector norm
norm: 제곱 합하고 루트씌우기

```python
v1=[1,2,3]
v1_sum=0
for v1_value in v1:
    v1_sum +=v1_value**2
v1_sum **=0.5
print(v1_sum)
```

Unit vector만들기

```python
#norm을 구하고 자기자신을 norm으로 나누기
#크기 합은 1이 된다.
''''
여러줄 가능
'''

v1=[1, 2, 3]
norm=0
for v1_values in v1:
    norm +=v1_values**2
norm **=0.5
for idx in range(len(v1)):
    v1[idx]/=norm
print(v1)
norm1=0
for v1_values in v1:
    norm1+=v1_values**2
print(norm1)
```

1-56 dot product하기

```python
v1, v2=[1,2, 3], [3,4,5]
dot_prod=0
for idx in range(len(v1)):
    dot_prod+=v1[idx]*v2[idx]
print(dot_prod)
```

유클리드 거리 구하기
2개의 리스트를 접근 하려면: 2개를 접근

```python
v1, v2=[1, 2, 3], [3, 4, 5]
diff=0
for idx in range(len(v1)):
    diff +=(v1[idx]-v2[idx])**2
print("distance:",diff**0.5) #거리라서 루트 씌우기
```

```python
v1, v2= [1, 2, 3],[3, 4, 5]
diff=0
for idx in range(len(v1)):
    diff+=(v1[idx]-v2[idx])**2
norm= diff**0.5
print(norm)
```

MSE 구현 뺴고 제곱하고 합의 평균! 유클리드거리는 루트

```python
predictions =[10, 20, 30]
labels= [10, 25, 40]
diff=0
for idx in range(len(predictions)):
        diff +=(predictions[idx]-labels[idx])**2
print("MSE Error:",diff/len(predictions))
```

1-59 다른 방법
numbers = [0, 2, 4, 2, 3, 1, 2, 3, 4, 2, 4]
이걸 [0, 0, 0, 0, 0]이란 그래프를 이용하여 위의 0부터 4까지 세어보기

```python
numbers=[0, 2, 4, 2, 3, 1, 3, 2, 4, 1, 2, 3, 1, 3, 4, 1, 2, 2, 2, 1, 4, 4]
count=[0 for _ in range(5)]
for idx in numbers:
    count[idx] +=1
print(count)
#numbers가 count의 인덱스가 되어서 count의 인덱스에있는 숫자에 접근해여 1씩 늘린다.
```
