1-32 norm은 제곱합을 하고 루트 씌우기

```python
v1=[1, 2, 3]
norm=(v1[0]**2+v1[1]**2+v1[2]**2)**0.5
print(norm)
```

```python
#다른방법
norm=0
norm+=v1[0]**2
norm+=v1[1]**2
norm+=v1[2]**2
norm**=0.5
print(norm)
```

```python
v1[0]/=norm
v1[1]/=norm
v1[2]/=norm
norm=(v1[0]**2+v1[1]**2+v1[2]**2)**0.5
print(norm)
```

1-35 MSE 사용

```python
predictions= [10, 25, 40]
labels= [10, 20, 30]
n_data=len(predictions)
mse=0
mse+=(labels[0]-predictions[0])**2
mse+=(labels[1]-predictions[1])**2
mse+=(labels[2]-predictions[2])**2
mse/=n_data
print(mse)
```

```python
mse=[]
mse.append((labels[0]-predictions[0])**2)
mse.append((labels[1]-predictions[1])**2)
mse.append((labels[2]-predictions[2])**2)
mse_sum=mse[0]+mse[1]+mse[2]
mse_result=mse_sum/len(mse)
print(mse_result)
```
