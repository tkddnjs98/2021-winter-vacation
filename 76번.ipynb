Sorting을 구현하기 내림 차순 정렬

1. scores라는 원래 리스트에서 최대값과 최소값 그리고 각각의 인덱스를 구하기 
2. 정렬한걸 표현하는 새로운 빈리스트 생성
3. 점수의 인덱스와 최대값의 인덱스와 같으면? 이때! 정렬을 위한 리스트에 append하기 
4. 나머지는 또 다른 임시의 리스트에 저장 
5. 마지막에 원래 리스트에 임시 리스트를 덮어 씌운다.
6. 위 과정을 반복한다.

```python
scores= [40, 20, 30, 10, 50]

sorted_list=[]
for _ in range(len(scores)):
    Max_idx,M=0,None #최대값의 인덱스와 최소값의 인덱스를 매번 0과 None으로 초기화
    for idx in range(len(scores)):
        if M==None or M<scores[idx]:
            M=scores[idx]
            Max_idx=idx 
    tmp_list=[] #tmp_list매번 초기화
    for idx in range (len(scores)):
        if idx ==Max_idx:
            sorted_list.append(scores[idx])
        else:
            tmp_list.append(scores[idx])
    print(scores)
    scores=tmp_list
#이경우 None이 되는 지 먼저 확인하고 Pass한다음 제일 큰 값의 여부를 확인한다. 
```
